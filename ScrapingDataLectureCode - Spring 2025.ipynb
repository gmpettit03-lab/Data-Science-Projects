## Using requests to collect files
Not all file types can be read directly into a dataframe. For those files we will have to download them first, then transform them to a format that we can read into a data frame. To download the files we will use the request package. 

```python
import requests 

r = requests.get('http://www.utk.edu')
print(r.status_code)
print(r.status_code == requests.codes.ok)
print(r.text)
```

# Scraping data from API's

API's (Application Programming Interface) are the modern way programs interact with each other. For data scientists the most common application are REST API's that share data in JSON (JavaScript Object Notation) form through a URL. We are going to start with [this](https://api.exchangeratesapi.io/latest?symbols=USD,GBP) simple API that reports excahnge rates for the Euro against a basket of currencies. This version requests USD and GBP. Hit the link to see the JSON that the API returns in your browser. 



```python
import requests
import pandas as pd

url = 'https://api.exchangerate.host/latest?base=USD&symbols=EUR,JPY,GBP,BTC,CAD&format=csv'
print(url)
df=pd.read_csv(url)
print(df)       
```
