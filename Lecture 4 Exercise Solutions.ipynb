```python
# Exercise 3: Create a function that takes an integer input and prints that element in the Fibonacci sequence.
```

```python
# This solution prints the full recursion sequence

def Fib(x):
    print("At recusion level ",x)
    if x==1:
        return 1
    if x==0:
        return 0
    else: 
        return Fib(x-1)+Fib(x-2)
```

```python
print("The Fibonacci sequence is: ", Fib(5))
```

```python
# This version includes two inputs to print the recursion level the first time through only

def Fib(x, seen):
    # If this number is smaller than what we've seen, print it
    if x < seen:
        print(f"Calculating Fibonacci for input: {x}")
        seen = x
        
    if x == 1:
        return 1, seen
    if x == 0:
        return 0, seen
    else:
        fib1, seen = Fib(x-1, seen)
        fib2, seen = Fib(x-2, seen)
        return fib1 + fib2, seen
```

```python
n=int(input("Enter a number and I will return that element of the Fibonacci sequence. "))
print("The value of the Fibonacci sequence's ", n, "element is: ", Fib(n, n+1))
```
