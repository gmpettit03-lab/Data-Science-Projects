## Defining and calling functions
Functions are chunks of code that we write and want to use multiple times in the same workbook, or across workbooks. Instead of copying and pasting code from one place to another, we can define a function. The function is defined with the "def" command. That's the code we want to run multiple times. After you have defined a function, you can call it using its name and any inputs it requires. 

```python
def absolute_value(x):     # define a function named 'abs' of one argument named 'x'
  if x >= 0:    # function body starts here
    return x
  else:
    return -x   # function body ends here
```

```python
# Test our new function here
absolute_value(0)
print(absolute_value(5))
print(absolute_value(-8))
```

The Fibonacci Sequence is a series of numbers where each number is the sum of the two preceding ones. In mathematical terms, it's defined as F(n) = F(n-1) + F(n-2), with base cases of F(0) = 0 and F(1) = 1. That creates a sequence which begins 0, 1, 1, 2, 3, 5, 8, 13, 21, and goes to infinity. This sequence appears frequently in nature, from the arrangement of leaves on plants to the spiral patterns of shells. It has mathematical properties that make it useful in fields from art to computer science.

```python
# Exercise: Create a function that takes an integer input and prints that element in the Fibonacci sequence

def Fib(x):
    if x==1:
        return 1
    if x==0:
        return 0
    else: 
        return Fib(x-1)+Fib(x-2)
    
```
