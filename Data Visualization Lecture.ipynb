# Data Visualization 
The representation of data in images. This is a broad category including bar graphs, maps, and infographics. We are going to focus mostly on graphs. The goal of this workbook is to cover a little bit of the theory of graphics then show you how to create graphs using the most common packages. 

## Edward Tufte
Edward Tufte, is a statistician and designer. His book, The Visual Display of Quantitative Information, is the primary reference for people who want to learn how to create simple, effective graphics. 

1. Comparisons:  Show data by comparisons (bar charts and the like) to depict contrasts and differences between dependent variables.

2. Causality:  Demonstrate how one or more independent variables impact or influence dependent variables.

3. Multivariate: Various data are combined so an audience can easily interpret an otherwise complex narrative.

4. Integration:  Incorporate various modes of information (texts, maps, calculations, diagrams, etc.), to show evidence of source data-to-findings. 

5. Documentation:  For credibility, include attribution, detailed titles, and measurements (scales).

6. Context: Describe or depict the before and after state. Show trend lines to hint at results in the future.


A simple way to evaluate graphs is the data to ink ratio. Compare the ink representing data compared to total ink in the graphic.

## Matplotlib
Matplotlib is a high level package with many basic graphing functions. Other visulization packages will use matplotlib tools. Keeping these packages updated and synced can be hard, but it allows us to use all the features that the higher level packages provide. Matplotlib is mostly closely integrated with Numpy a package with math and statistics functions. 

```python
import matplotlib.pyplot as plt
import numpy as np


# make data
x = np.linspace(0, 10, 100)     # create a numpy array with 100 numbers evenly spaced between 0 and 10
y = 4 + 2 * np.sin(2 * x)       # Create a sine wave 

# plot
fig, ax = plt.subplots()        # Create a graphic with a figure and imgaes
ax.plot(x, y)                   # Plot x against y
```

### Defining a graph in matplotlib
What's goign on with the command fig, ax=plt.subplots? Matplot lib allows you to define the figure and axes separately. You might want to put multiple axes on one figure. Matplotlib gives you the flexibility to do that. Let's look at the axis and figure from the plot above.



```python
# The axis part lets you describe the range of the plot
print(ax)
print(type(ax))
# The figure part lets you define the area of the graph. You can re-size, but not draw
print(fig)
print(type(fig))

# We can put two axes in one figure:
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 5))       # subplots defined as row, columns 
fig.tight_layout(pad=3)
plt.show()

# We can use the figure commands to change the size of the layout
plt.figure(figsize=(10,6))
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')                          # Give our figure a different style#
#plt.style.use('dark_background')                                 # Give our figure a different style
#plt.style.use('bmh')                                     # Give our figure a different style
#plt.style.use('darkbackround')                           # Give our figure a different style
#plt.style.use('ggplot')                                  # Give our figure a different style


# Create a data set
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# Now make the graph
fig, ax = plt.subplots()                                # Define a graph by it's data (fig) and axes

ax.plot(x, y, linewidth=2.0)                            # Make the line thicker                           

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8, 2))          # Define axis length and tick frequency

# We can add titles and labels to help 
plt.title("Plot of X against Y")
plt.xlabel('X axis ')
plt.ylabel('Y axis ')


plt.show()                                              # Print the graph to the Jupyter notebook
```

```python
# We can create 3d plots by defining a 3rd axis. 

z= 6 + .85 * np.sin(2 * x)
# create a third variable. 
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x, y, z)

plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")
df.describe()

# Show the results of a linear regression within each dataset
sns.lmplot(x="x", y="y", col="dataset",  hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
               palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```

```python

from plotnine.data import mtcars       # Use the MTcars data set that has info from 1972 car and driver road tests
import pandas as pd

df=pd.DataFrame(mtcars)
print(df.describe())

print(df.head(2))
```

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg'))
 + geom_point())
```

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))
```
