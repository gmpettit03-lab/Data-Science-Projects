# Debugging
When a piece of code does not produce the expected result, it needs to be debugged. That is the process of reviewing the code to find the problem and correcting it. Debugging can be frustrating. It is often hard to see simple mistakes you just made. Your mind reads what you wanted to type, not what you did type. There are a couple of tips for debugging, depending on the type of error:

1) Synatx errors: If your code crashes before completing, then you likely have a syntax error somewhere. Copy the code to a clean environment and execute a few lines at a time. Keep going until the program crashes. Once you have isolated the line that is causing the problem you can see if identify the source of the error. You may have to fix a bug somewhere above the line that causes the crash, but you know the bug is not below the last line to run successfully.

2) Logic errors: These tend to produce either incorrect or inconsistent results. If your program works some of the time, but not others, there may be a logic error. If the program is complicated enough it makes sense to diagram it with paper and pencil. Make sure you understand which parts of the code should be executed under different conditions. Then use print statements to make sure that part of the code is in fact running. You can print, "This statement should print if X>5" and then make sure that when x is bigger than 5 you see that statement on the terminal. For these types of bugs, finding the conditions that create the error is extremely important. What causes the program to work? What causes it to fail? Identifying these conditions can help narrow down the location of the bug.

3) Runtime errors: Your code runs, but produces an unexpected output. This could be caused by runtime errors, logic errors or both. Adding print statements can be helpful here. Print the value of your variables every time you update them. Check to make sure those values are what you are expecting. Printing something like, "The value of x is {x}." before and after every expression that changes the value of x can help you track what is going on throughout the program and identify the source of the bug. 



## Debugging vs. Testing
Two separate tasks:
1) Debugging: Fix code that does not run 
* Python provides (occasionally useful) error messages to help with de-bugging 
* Read error messages carefully 
* Look at code just above the error 
* Take a walk and clear your mind 

2) Testing: Check that code produces correct output 
* Testing is harder 
* Test often and modularly  
* Try to break the code 
* Use subsets of the data until the code is working
